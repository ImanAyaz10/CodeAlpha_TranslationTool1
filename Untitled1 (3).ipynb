```python
from groq import Groq
import gradio as gr
import os
from google.colab import userdata

client = Groq(api_key=userdata.get("GROQ_API_KEY"))
```

```python
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": "Translate Hello to Korean"}],
)

print(response.choices[0].message.content)
```

```python
def translate_text(text, source_lang, target_lang):

    if text.strip() == "":
        return "Please enter some text."

    prompt = f"""
    Translate the following text from {source_lang} to {target_lang}.
    Only return the translated text.

    Text: {text}
    """

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )

    return response.choices[0].message.content.strip()


iface = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(lines=4, placeholder="Enter text here..."),
        gr.Dropdown(["English", "Urdu", "Korean", "French", "German"], label="Source Language"),
        gr.Dropdown(["English", "Urdu", "Korean", "French", "German"], label="Target Language")
    ],
    outputs=gr.Textbox(label="Translated Text"),
    title="🌍 AI Language Translator (Groq + LLaMA 3.3)",
    description="Translate text between languages using Groq's LLaMA 3.3 model"
)

iface.launch()
```
